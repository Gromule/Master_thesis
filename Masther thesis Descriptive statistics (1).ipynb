```python
import pandas as pd

# Function to filter out two-digit values from a column without decimal places
def filter_two_digits(value):
    if pd.notnull(value) and isinstance(value, float) and value.is_integer():
        value = int(value)
    if pd.notnull(value) and isinstance(value, int) and 10 <= value <= 99:
        return value
    else:
        return None

# Apply the function to filter two-digit values from both columns
data['Today_age'] = data['Calculated_Age'].apply(filter_two_digits)
data['Today_age'] = data['Today_age'].fillna(data['Calculated_Age_2'].apply(filter_two_digits))

# Remove the decimal part from the values in the 'Today_age' column
data['Today_age'] = data['Today_age'].astype(float).astype('Int64')
```
